# CHEMBL FG comparison

```python
import pandas as pd
from collections import Counter
chembl_efg_fgs_df = pd.read_csv('molecule_data/chembl_efg_fgs.csv')
chembl_accfg_fgs_df = pd.read_csv('molecule_data/chembl_approved_sm_fgs.csv')
chembl_accfg_fgs_df_cleaned = chembl_accfg_fgs_df[chembl_accfg_fgs_df['smiles'].isin(chembl_efg_fgs_df['smiles'])]
chembl_accfg_count = Counter()
for fgs_dict in chembl_accfg_fgs_df_cleaned['fgs']:
    fgs = eval(fgs_dict)
    fgs_count = {fg: len(fg_list) for fg, fg_list in fgs.items()}
    chembl_accfg_count.update(fgs_count)
total_count = sum(chembl_accfg_count.values())    
fgs_count_df = pd.DataFrame(chembl_accfg_count.items(), columns=['Functional Group', 'Count'])
fgs_count_df = fgs_count_df.sort_values(by='Count', ascending=False).reset_index(drop=True)
fgs_count_df['Percentage'] = fgs_count_df['Count']/total_count*100
fgs_count_df.to_csv('molecule_data/chembl_accfg_fg_dist.csv', index=False)
```
